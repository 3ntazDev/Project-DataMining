```python
import pandas as pd
import re 
df = pd.read_csv('cleaned_data.csv')
df['phone_1'] = df['phone_1'].str.replace(')', '', regex=False)

def format_phone_number(phone, index):
    if index in [32, 40]:  
        return phone
    if isinstance(phone, str) and phone.startswith('+966'):
        digits = re.sub(r'\D', '', phone[4:]) 
        formatted_phone = f"+966 {digits[:2]} {digits[2:5]} {digits[5:9]}"
        return formatted_phone
    return phone

df['phone_1'] = df['phone_1'].apply(lambda x: x.strip() if isinstance(x, str) else x)  
df['phone_1'] = df.apply(lambda row: format_phone_number(row['phone_1'], row.name), axis=1)
df.replace({0: 'Not provided', 'Nothing': 'Not provided'}, inplace=True)


display(df)

```
